```python
import torch
from torch_geometric.data import Data


edge_index = torch.tensor([[0, 1, 1, 2,3],
                           [1, 0, 2, 1,1]], dtype=torch.long) 

data = Data(edge_index=edge_index)
data.to_dict()
```

```python
from torch_geometric.utils.convert import to_networkx
import matplotlib.pyplot as plt
import networkx as nx
vis = to_networkx(data)
plt.figure(1,figsize=(8,8)) 
nx.draw(vis, cmap=plt.get_cmap('Set3'),node_size=70,linewidths=6)
plt.show()
```

```python
#delete node 

def delete_edge(N, data):
    edgeIndices= data.edge_index;
    listFrom, listTo = edgeIndices[0], edgeIndices[1]
    newListFrom, newListTo = [], []
    for i in range(len(listFrom)):
        if (listFrom[i] == N or listTo[i] == N):
            continue;
        newListFrom.append(listFrom[i])
        newListTo.append(listTo[i])
    data1 =  Data(edge_index=torch.tensor([newListFrom, newListTo], dtype=torch.long));
    data1.num_nodes = data.num_nodes 
    return data1;

data = delete_edge(2, data);

```

```python
heh = torch.tensor([0,1,2], dtype=torch.long)
data = data.subgraph(heh)
```

```python
vis = to_networkx(data)
plt.figure(1,figsize=(8,8)) 
nx.draw(vis, cmap=plt.get_cmap('Set3'),node_size=70,linewidths=6)
plt.show()
```

```python
from torch_geometric.datasets import KarateClub
karate = KarateClub()
```
